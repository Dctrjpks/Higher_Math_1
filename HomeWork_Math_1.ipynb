```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x1=np.linspace(-3.14,3.14,51)
```

```python
y1 = np.cos(x1)
y2 = np.cos(2 * x1)
y3 = np.cos(.5 * x1)
```

```python
plt.plot(x1, y1, color = 'red', linestyle = 'solid', label = 'cos(x)')
plt.plot(x1, y2, color = 'green', linestyle = 'dashed', label = 'cos(2x)')
#plt.plot(x1, y3, color = 'green', linestyle = 'dashed',label = 'cos(1/2x)')
```
